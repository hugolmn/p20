# Générateur d'une variable aléatoire uniforme sur [0,1]

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare
```

```python
# Définition du nombre de valeurs et du nimbre d'intervalles
N = 1000
r = 15
```

```python
# Réalisations de variables suivant la loi uniforme sur [0,1]
x = np.random.uniform(0, 1, N)
```

```python
# Affichage des réalisations
plt.scatter(np.arange(len(x)), x) 
```

```python
# Histogramme sur 10 intervalles
plt.hist(x, bins=r)
plt.show()
```

```python
# Calcul de K_n
K_n = sum((np.histogram(x, bins=r)[0] - N/r)**2 / (N/r))
K_n
```

```python
# Calcul de K_n
chisquare(
    f_obs=np.histogram(x, bins=r)[0],
    f_exp=N/r,
    ddof=r-1, 
)
```
