```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
x1 = np.random.uniform(0,1,10000)
x2 = np.random.uniform(0,1,10000)
```

```python
plt.hist(x1)
plt.show()
```

```python
plt.hist(x2)
plt.show()
```

```python
z1 = np.sqrt(-2*np.log(x1)) * np.cos(2 * np.pi * x2)
z2 = np.sqrt(-2*np.log(x1)) * np.sin(2 * np.pi * x2)
```

```python
mu = 0
variance = 1
sigma = np.sqrt(variance)
x = np.linspace(-4, 4, 1000)
```

```python
plt.hist(z1, bins=40, alpha=.5, density=True)
plt.hist(z2, bins=40, alpha=.5, density=True)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```
